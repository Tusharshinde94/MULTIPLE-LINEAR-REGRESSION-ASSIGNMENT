```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
data=pd.read_csv("D:/ASSIGNMENTS/Assign5(multiple Regression)/50_Startups.csv")
data.head()
```

```python
data1=data.rename({'R&D Spend':'RDS','Administration':'ADMIN','Marketing Spend':'MARKET','State':'STATE'},axis=1)
data1.head(2)
```

```python
model=smf.ols("Profit~RDS+ADMIN+MARKET",data=data1).fit()
```

```python
model.params
```

```python
model.tvalues,'\n',np.round(model.pvalues,3)
```

```python
ml_a=smf.ols("Profit~ADMIN",data=data1).fit()
ml_a.tvalues,'\n',np.round(ml_a.pvalues,3)
```

```python
ml_m=smf.ols("Profit~MARKET",data=data1).fit()
ml_m.tvalues,'\n',np.round(ml_m.pvalues,3)
```

```python
ml_am=smf.ols("Profit~ADMIN+MARKET",data=data1).fit()
ml_am.tvalues,'\n',np.round(ml_am.pvalues,3)
```

```python
model.rsquared,model.rsquared_adj
```

```python
qqplot=sm.qqplot(model.resid,line='q')
```

```python
def standard_values(vals):
    return(vals-vals.mean())/vals.std()
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
```

```python
fig=plt.figure(figsize=(16,7))
fig=sm.graphics.plot_regress_exog(model,'RDS',fig=fig)
```

```python
fig=plt.figure(figsize=(16,7))
fig=sm.graphics.plot_regress_exog(model,'ADMIN',fig=fig)
```

```python
fig=plt.figure(figsize=(16,7))
fig=sm.graphics.plot_regress_exog(model,'MARKET',fig=fig)
```

```python
influence_plot(model)
plt.show()
```

```python
(c,_)=model.get_influence().cooks_distance
c
```

```python
fig=plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(data1)),np.round(c,3))
plt.title('Cook Distance Graph')
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```
